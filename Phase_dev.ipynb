```python
import skrf as rf
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d
```

```python
HPol_path = 'Data/Port1inNWhole_Port2inShole_ST35_HPOL_VNA'
VPol_path = 'Data/Port1inNWhole_Port2inShole_ST35_VPOL_VNA'
s2p_files_HPol = glob.glob(os.path.join(HPol_path, '*.s2p'))
s2p_files_VPol = glob.glob(os.path.join(VPol_path, '*.s2p'))
# Load all s2p files into a single list of Network objects
networks = [rf.Network(file) for file in s2p_files_HPol + s2p_files_VPol]
```

```python
# networks[0].plot_s_deg(0,1)
dir = 'Phase'
network = networks[0]

f_low = 200
f_high = 300

plt.figure(figsize=(40, 6))
plt.plot(network.f / 1e6, network.s_deg[:, 1, 0], label='S12 Phase')
# plt.plot(network.f, network.s_deg[:, 1, 0], label='S12 Phase')

# plt.plot(network.f / 1e9, network.s_deg[:, 0, 0], label='S11 Phase')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Phase (degrees)')
plt.xlim(f_low, f_high)
plt.title(f'Phase Response of {network.name} S21')
plt.grid(True)
plt.legend()
plt.savefig(f'{dir}/{network.name} S21 phase test.png', bbox_inches='tight', dpi=300)
```
